# Check that how many MBs are there in one KB? and how to calculate?
1. Convert KBs into MBs
2. Then remove M from all numbers
3. Handle 'varies with device' correctly.

```python
df['Size'].isnull().sum()
```

```python
# convert Size column to numeric
df['Size'] = df['Size'].apply(lambda x: str(x).replace('M','') if 'M' in str(x) else str(x))
df['Size'] = df['Size'].apply(lambda x: str(x).replace('k','') if 'k' in str(x) else str(x))
df['Size'] = df['Size'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
df['Size'] = df['Size'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
```

```python
# convert 'varies with device' in Size column to Null values using numpy package
df['Size'] = df['Size'].apply(lambda x: np.nan if 'Varies with device' in str(x) else str(x))
```

```python
# convert Size column to Megabytes (All kilobytes to megabytes)
df['Size_MB'] = df['Size'].apply(lambda x: (float(x) / 1024)/1024)
```

# do it as numeric column or group that into different groups/binnings

```python
# make binning of Installs column
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
bins = [0, 1000, 10000, 100000, 1000000, 10000000]
labels = ['0-1K', '1K-10K', '10K-100K', '100K-1M', '1M-10M']
df['Installs_binned'] = pd.cut(df['Installs'], bins=bins, labels=labels)

print(df['Installs_binned'])
```
